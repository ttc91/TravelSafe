```python
def sigmoid(s):
    return 1 / (1 + np.exp(-s))
```

```python
def prod(w, X):
    '''
    X: a 3d array of shape (N, d). N datapoint, each with size d
    w: a 1d numpy array of shape (d)
    '''
    return sigmoid(X.dot(w))
```

```python
# refer to: formulation/logistic/loss_function.png
def loss(w, X, y, lamda):
    z = prod(w, X)
    return - np.mean(y * np.log(z) + (1 - y) * np.log(1 - z) + 0.5 * lamda * np.sum(w[1:] ** 2))
```

```python
def logistic_sigmoid_regression(w_init, X, y, lamda = 0.0001, learning_rate = 0.05, epochs = 1000):
    w_list = [w_init]
    loss_history = [loss(w_init, X, y, lamda)]
    epoch = 0
    n = X.shape[1]
    d = X.shape[0]
    while epoch < epochs:
        mix_idx = np.random.permutation(n)
        for i in mix_idx:
            # size of xi is (3, 1)
            xi = X[i,:]
            yi = y[i]
            zi = prod(w_list[-1].T, xi)
            w = w_list[-1] - learning_rate * ((zi - yi) * xi + lamda * w_list[-1])
            w_list.append(w)
            if np.linalg.norm(w_list[-1] - w_list[-2]) / d < 1e-6:
                break
            epoch += 1
        loss_history.append(loss(w_list[-1], X, y, lamda))
    return w_list, loss_history
```

```python
w_init = np.random.randn(X_train.shape[1])
w_init
```
